```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
akida = pd.read_csv('../frameworks/akida/notebooks/system_monitoring_data_akida.csv')
loihi = pd.read_csv('..//frameworks/lava/notebooks/system_monitoring_data_loihi.csv')

# For fair comparison, align time axes if needed (optional: interpolate or trim)
# Here, we assume both start at 0 and are comparable

# Plot CPU Usage
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(akida['Timestamp'], akida['CPU_Usage_Percent'], label='Akida', color='tab:blue')
plt.plot(loihi['Timestamp'], loihi['CPU_Usage_Percent'], label='Lava (Loihi)', color='tab:orange', alpha=0.7)
plt.xlabel('Time (s)')
plt.ylabel('CPU Usage (%)')
plt.xlim(0, max(akida['Timestamp'].max(), loihi['Timestamp'].max()))
plt.ylim(0, 100)
plt.title('CPU Usage Over Time')
plt.legend()
plt.grid(True)

# Plot GPU Usage
plt.subplot(1, 2, 2)
plt.plot(akida['Timestamp'], akida['GPU_Usage_Percent'], label='Akida', color='tab:blue')
plt.plot(loihi['Timestamp'], loihi['GPU_Usage_Percent'], label='Lava (Bootstrap)', color='tab:orange', alpha=0.7)
plt.xlabel('Time (s)')
plt.ylabel('GPU Usage (%)')
plt.xlim(0, max(akida['Timestamp'].max(), loihi['Timestamp'].max()))
plt.ylim(0, 100)
plt.title('GPU Usage Over Time')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
